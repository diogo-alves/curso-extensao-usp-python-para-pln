### Usando bigramas e trigramas para encontrar entidades nomeadas

```python
import nltk
from nltk import bigrams, trigrams


with open(CORPUS_TXT) as f:
    corpus = f.read()
    tokens = nltk.word_tokenize(corpus)
    bigrams_tokens = list(bigrams(tokens))
    trigrams_tokens = list(trigrams(tokens))
    
    for word1, word2 in bigrams_tokens:
        if word1[0].isupper() and word2[0].isupper():
            print(f'entidade: {word1} {word2}')

    for word1, word2, word3 in trigrams_tokens:
        if word1[0].isupper() and word2[0].isupper() and word3[0].isupper():
            print(f'entidade: {word1} {word2} {word3}')
```

## Stemização e Lematização


*   Stemização - reduz uma palavra ao seu radical (independe de regra gramatical): amigo, amiga, amigão -> amig
*   Lematização - reduz a palavra à sua forma canônica: propôem, propuseram, propondo -> propor

```python
import nltk 
 

# exemplo de stemização com RSLP (Removedor de Sufixos da Lingua Portuguesa)
stemmer = nltk.RSLPStemmer()
words = 'amigo amiga amigão propuzeram propõem propondo'.split()
for word in words:
    print(f'Stemização da palavra {word!r}: {stemmer.stem(word)!r}')

# como não existe lematizador em pt-BR no nltk, apenas palavras em inglês são lematizadas
lemmatizer = nltk.stem.WordNetLemmatizer()
words = 'propõem estudei studied studying sings'.split()
for word in words:
    print(f"Lematização da palavra {word!r}: {lemmatizer.lemmatize(word, pos='v')!r}")

```

## Etiquetadores

São responsáveis por identificar as classes gramaticais das palavras.

```python
import nltk
from nltk.corpus import mac_morpho
from nltk.tag import DefaultTagger, UnigramTagger


with open(CORPUS_TXT) as f:
    corpus = f.read()
    tokens = nltk.word_tokenize(corpus)
    trainning_sents = mac_morpho.tagged_sents()
    # palavras que não existem nas sentenças de treino serão etiquetadas com este valor padrão
    default_tagger = DefaultTagger('N')
    tagger = UnigramTagger(trainning_sents, backoff=default_tagger)
    tagged_tokens = tagger.tag(tokens)
    print(tagged_tokens)
```

```python
from nltk.chunk import RegexpParser


pattern = 'NP: {<NPROP><NPROP>|<N><N>}'
parser = RegexpParser(pattern)
tree = parser.parse(tagged_tokens)
print(tree)
```

## Exercício de fixação

Faça uma análise descritiva completa do nosso corpus de teste utilizando as funções do NLTK.

Exemplos de atributos:


*   Quantidade de tokens
*   Quantidade de sentenças / média do tamanho das sentenças
*   Quantidade de substantivos, adjetivos, advérbios...
*   Quantidade de palavras com o mesmo radical
*   Quantidade de símbolos de pontuação
*   Palavras mais frequentes do corpus


```python
import string

import nltk
from nltk.corpus import mac_morpho
from nltk.tag import DefaultTagger, UnigramTagger


with open(CORPUS_TXT) as f:
    corpus = f.read()

    # quantidade de palavras no corpus
    tokens = nltk.word_tokenize(corpus)
    print('Quantidade de palavras no corpus: ', len(tokens))

    # Quantidade de sentenças
    sents = nltk.sent_tokenize(corpus)
    print('Quantidade de sentenças: ', len(sents))

    # Média do tamanho das sentenças
    tokens_per_sent = [len(nltk.word_tokenize(sent)) for sent in sents]
    avg_sents = sum(tokens_per_sent) / len(sents)
    print('Média de tamanho das sentenças: ', avg_sents)

    # Quantidade de substantivos, adjetivos, advérbios, etc
    trainning_sents = mac_morpho.tagged_sents()
    default_tagger = DefaultTagger('N')
    tagger = UnigramTagger(trainning_sents, backoff=default_tagger)
    tagged_tokens = tagger.tag(tokens)
    pos = nltk.FreqDist(tag for _, tag in tagged_tokens)
    print('Ocorrências de classes gramaticais: ', pos.most_common())

    # Quantidade de palavras com o mesmo radical
    stemmer = nltk.RSLPStemmer()
    word_stems = nltk.FreqDist(stemmer.stem(word) for word, _ in tagged_tokens)
    print('Quantidade de palavras com o mesmo radical: ', word_stems.most_common())

    # Quantidade de símbolos de pontuação
    punct_tokens = [token for token in nltk.wordpunct_tokenize(corpus) if token in string.punctuation]
    print('Quantidade de símbolos de pontuação: ', len(punct_tokens))

    # Palavras mais frequentes do corpus
    frequence = nltk.FreqDist(tokens)
    print('Palavras mais frequentes do corpus: ', frequence.most_common())
```

## Exemplo de uso com um trecho de um artigo do Meu Timão

```python
import spacy


nlp = spacy.load('pt_core_news_lg')
doc = nlp(
    'O goleiro Cássio quer jogar até os 40 anos de idade e, claro, espera que '
    'isso aconteça com a camisa do Corinthians. Foi isso que ele disse na tarde '
    'desta sexta-feira, no CT Joaquim Grava, ao ser lembrado que, caso entre na '
    'partida contra o Red Bull Bragantino, pelo Brasileiro, irá igualar o número '
    'de jogos de Claudio Christovam de Pinho, o maior artilheiro da história do clube.'
)

# no spacy os tokens são instâncias da classe Token
# usamos o atributo orth_ (ou text) para acessar a string correspondente ao token
tokens = [token.orth_ for token in doc]
print('Tipo do atributo: ', type(tokens[0]))

# é possível verificar os tipos de tokens através de properties como is_digit, is_alpha, is_punct, etc

digit_tokens = [token.orth_ for token in doc if token.is_digit]
print('Dígitos no texto: ', digit_tokens)

alpha_tokens = [token.orth_ for token in doc if token.is_alpha]
print('Palavras no texto: ', alpha_tokens)

punct_tokens = [token.orth_ for token in doc if token.is_punct]
print('Puntuações no texto: ', punct_tokens)
```

## 1. Escolha um corpus qualquer (pode ser o que foram utilizados na aula) e, usando as funções do NLTK, faça:

a. Extraia todos os bigramas do corpus

b. Extraia todos os trigramas do corpus

c. Extraia todos os 4-gramas do corpus

d. Retorne as entidades nomeadas do corpus, usando os bigramas e trigramas

e. Escolha 3 palavras do seu corpus e faça o stemming delas

f. Separe uma sentença do seu corpus e retorne todas as classes gramaticais das palavras da sentença. Analise se o etiquetador acertou todas as classes gramaticais.

g. Retorne as classes gramaticais de todas as palavras do seu corpus

h. Retorne as entidades nomeadas do seu corpus, usando a técnica de chunking

```python
corpus = (
    'O prêmio Nobel em Física do ano de 2021 foi dedicado à compreensão de sistemas complexos, '
    'dentre eles o que permite a compreensão das mudanças climáticas que afetam nosso planeta. '
    'Os pesquisadores Syukuro Manabe, dos Estados Unidos, e Klaus Hasselmann, da Alemanha, foram '
    'premiados especificamente por modelarem o clima terrestre e fazerem predições sobre o aquecimento '
    'global. A outra metade do prêmio foi para Giorgio Parisi, da Itália, que revelou padrões ocultos '
    'em materiais complexos desordenados, das escalas atômica à planetária, em uma contribuição essencial '
    'à teoria de sistemas complexos.'
)
```

```python
from string import punctuation

import nltk
from nltk.corpus import mac_morpho
from nltk.chunk import RegexpParser
from nltk.tag import DefaultTagger, UnigramTagger


tokens = nltk.word_tokenize(corpus)

print('a. Extraia todos os bigramas do corpus')
bigrams = list(nltk.bigrams(tokens))
print(bigrams, end='\n\n')

print('b. Extraia todos os trigramas do corpus')
trigrams = list(nltk.trigrams(tokens))
print(trigrams, end='\n\n')

print('c. Extraia todos os 4-gramas do corpus')
ngrams = list(nltk.ngrams(tokens, 4))
print(ngrams, end='\n\n')

print('d. Retorne as entidades nomeadas do corpus, usando os bigramas e trigramas')
ents = [' '.join(ngram) for ngram in bigrams + trigrams if all(word[0].isupper() for word in ngram)]
print(ents, end='\n\n')

print('e. Escolha 3 palavras do seu corpus e faça o stemming delas')
words = 'permite modelarem revelou'.split()
stemmer = nltk.RSLPStemmer()
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words, end='\n\n')

print(
    'f. Separe uma sentença do seu corpus e retorne todas as classes gramaticais das '
    'palavras da sentença. Analise se o etiquetador acertou todas as classes gramaticais.'
)
sent = nltk.sent_tokenize(corpus)[0]
training_sents = mac_morpho.tagged_sents()
default_tagger = DefaultTagger('N')
tagger = UnigramTagger(training_sents, backoff=default_tagger)
words = [token for token in nltk.word_tokenize(sent) if token not in punctuation]
tagged_words = tagger.tag(words)
print(tagged_words, end='\n\n')

print('g. Retorne as classes gramaticais de todas as palavras do seu corpus')
words = [token for token in tokens if token not in punctuation]
tagged_words = tagger.tag(words)
print(tagged_words, end='\n\n')

print('h. Retorne as entidades nomeadas do seu corpus, usando a técnica de chunking')
pattern = 'NP: {<NPROP>}'
parser = RegexpParser(pattern)
tree = parser.parse(tagged_words)
for subtree in tree.subtrees():
    if subtree.label() == 'NP':
        print(subtree)
```

## 2. Utilizando um corpus de seu interesse (pode ser o que foi utilizado na aula) e utilizando o spaCy, faça:

a. Tokenize todo o corpus

b. Tokenize todo o corpus e retorne a lista com strings

c. Tokenize todo o corpus e retorne apenas as palavras

d. Retorne a quantidade de palavras do corpus

e. Tokenize todo o corpus e retorne apenas os números

f. Retorne a quantidade de números do corpus

g. Tokenize todo o corpus e retorne apenas as pontuações

h. Retorne a quantidade de pontuações do corpus

i. Retorne a frequência de pontuação do corpus (obs: pode usar o NLTK nesse exercício!)

j. Tokenize todo o corpus e retorne a quantidade de espaços presentes no corpus

```python
import nltk
import spacy


nlp = spacy.load('pt_core_news_lg')
doc = nlp(corpus)

print('a. Tokenize todo o corpus')
tokens = [token for token in doc]
print(tokens, end='\n\n')

print('b. Tokenize todo o corpus e retorne a lista com strings')
tokens = [token.text for token in doc]
print(tokens, end='\n\n')

print('c. Tokenize todo o corpus e retorne apenas as palavras')
tokens = [token.text for token in doc if token.is_alpha]
print(tokens, end='\n\n')

print('d. Retorne a quantidade de palavras do corpus')
print(len(tokens), end='\n\n')

print('e. Tokenize todo o corpus e retorne apenas os números')
tokens = [token.text for token in doc if token.is_digit]
print(tokens, end='\n\n')

print('f. Retorne a quantidade de números do corpus')
print(len(tokens), end='\n\n')

print('g. Tokenize todo o corpus e retorne apenas as pontuações')
tokens = [token.text for token in doc if token.is_punct]
print(tokens, end='\n\n')

print('h. Retorne a quantidade de pontuações do corpus')
print(len(tokens), end='\n\n')

print('i. Retorne a frequência de pontuação do corpus (obs: pode usar o NLTK nesse exercício!)')
frequence = nltk.FreqDist(tokens)
print(frequence.most_common(), end='\n\n')

print('j. Tokenize todo o corpus e retorne a quantidade de espaços presentes no corpus')
tokens = [token.whitespace_ for token in doc if token.whitespace_]
print(len(tokens))
```

## 3. Ainda utilizando o mesmo corpus e o spaCy, faça:

a. Analise o 1º parágrafo do corpus, lematizando todos os tokens possíveis. Verifique se foi retornado os lemas corretos.

b. Ainda com o 1º parágrafo, retorne todas as classes gramaticais dos tokens.

c. Retorne todas as classes gramaticais do corpus.

d. Use o 1º parágrafo para retornar as informações morfológicas dos tokens.

e. Retorne todas as entidades nomeadas do corpus.

f. Verifique a classificação de cada entidade nomeada do corpus.

g. Use o displaCy para mostrar graficamente todas as entidades nomeadas do corpus.

```python
import nltk
import spacy
from IPython import display


nlp = spacy.load('pt_core_news_lg')
doc = nlp(corpus)
first_paragraph = next(doc.sents)

print('a. Analise o 1º parágrafo do corpus, lematizando todos os tokens possíveis. Verifique se foi retornado os lemas corretos.')
lemmas = [(token.text, token.lemma_) for token in first_paragraph if token.text != token.lemma_]
print(lemmas, end='\n\n')

print('b. Ainda com o 1º parágrafo, retorne todas as classes gramaticais dos tokens.')
tokens = [(token.text, token.pos_) for token in first_paragraph]
print(tokens, end='\n\n')

print('c. Retorne todas as classes gramaticais do corpus.')
tokens = [(token.text, token.pos_) for token in doc]
print(tokens, end='\n\n')

print('d. Use o 1º parágrafo para retornar as informações morfológicas dos tokens.')
tokens = [(token.text, token.morph) for token in first_paragraph]
print(tokens, end='\n\n')

print('e. Retorne todas as entidades nomeadas do corpus.')
ents = [ent for ent in doc.ents]
print(ents, end='\n\n')

print('f. Verifique a classificação de cada entidade nomeada do corpus.')
ents = [(ent, ent.label_) for ent in doc.ents]
print(ents, end='\n\n')

print('g. Use o displaCy para mostrar graficamente todas as entidades nomeadas do corpus.')
html_code = spacy.displacy.render(doc, style='ent')
display.HTML(html_code)
```
